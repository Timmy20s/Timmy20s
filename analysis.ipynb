```python
import pandas as pd

# Load the dataset
df = pd.read_csv("HollywoodsMostProfitableStories.csv")

# Display first 5 rows
df.head()
```

```python
df['Genre'].value_counts().plot(kind='bar', figsize=(10,5), title='Number of Films by Genre')
```

```python
df.groupby('Genre')['Profitability'].mean().sort_values(ascending=False).plot(
    kind='bar', figsize=(10,5), title='Average Profitability by Genre'
)
```

```python
import matplotlib.pyplot as plt

plt.scatter(df['Audience  score %'], df['Profitability'])
plt.xlabel('Audience Score (%)')
plt.ylabel('Profitability')
plt.title('Does Audience Score Affect Profitability?')
plt.show()
```
